```python
df.drop_duplicates(inplace=True)
```

```python
# outliers detection

for i in df.columns:
    sns.boxplot(df[i])
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.show()
```

```python
# outliers removal
out_list = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']

for i in out_list:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    
    IQR = Q3-Q1
    
    df = df[(df[i]>=Q1-1.5*IQR) & (df[i]<=Q3+1.5*IQR)]
```

```python
for i in df.columns:
    sns.boxplot(df[i])
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.show()
```

```python
# define X and y

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
l_model = LinearRegression()
```

```python
# training the model
l_model.fit(x_train,y_train)
```

```python
# prediction
prediction = l_model.predict(x_test)
```

```python
prediction
```

```python
from sklearn.metrics import *
```

```python
r2score = r2_score(y_test, prediction)
```
