```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area'  # Replace with the URL of the web page
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
table_data = []

for row in table.find_all('tr'):  # Loop through each row in the table
    columns = row.find_all('td')   # Find all the columns in the row
    row_data = [column.get_text(strip=True) for column in columns]  # Extract text from each column
    table_data.append(row_data)
```

```python
table_data
```

```python
df = pd.DataFrame(table_data, columns=['Rank','Country', '%Total','Area in km2',' '])  # Replace with your column names
```

```python
df
```

```python
# Remove the first row (index 0)
df = df.drop(0)


# Remove the last row (use -1 to refer to the last row)
df = df.drop(df.index[-1])
```

```python
df
```

```python
# Remove the last column
df = df.drop(df.columns[-1], axis=1)
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
x=df['Country']
y=df['%Total']
z=df['Area in km2']
```
