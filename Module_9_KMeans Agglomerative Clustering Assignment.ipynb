## Tasks to be performed:
- Load iris data from load_iris function from sklearn.datasets package.
- From the dataset extract the data property.
- Train an AgglomerativeClustring model based on the data.
- Plot dendrogram to visualize the clustering linkage

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Iris-1.csv')
```

```python
X = data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = data[['Species']]

X = X.to_numpy()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
pred_cluster = model.fit_predict(X)
```

```python
data['cluster']= pred_cluster
```

```python
data['cluster'].unique()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
plt.rcParams['figure.figsize'] = [20,15]
Z = linkage(X,'ward')
dendrogram_plot = dendrogram(Z, truncate_mode = 'level')
plt.show()
```
